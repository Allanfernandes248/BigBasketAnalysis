```python
df['discount_amount'] = df['market_price'] - df['sale_price']
```

```python
discounted_dataframe = df[df['discount_amount'] > 0]
discounted_dataframe
```

```python

#discount formula --> ((market price - sale prioce)/market price ) * 100 
# creating a discount percentage column 
discounted_dataframe['discount_percentage'] = ((df['market_price'] - df['sale_price'])/df['market_price'])*100
```

```python
discounted_dataframe
```

h1# rounding_off the percetage upto 2 decimal places 


```python

converter=discounted_dataframe['discount_percentage']
# discounted_dataframe['discount_percentage'] = round(df['discount_percentage'],2)
converter = round(converter,2)
discounted_dataframe['discount_percentage'] = converter 
```
